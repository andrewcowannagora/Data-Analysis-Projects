```python
from pycaret.utils import version
version()
```

```python
# import libraries
from pycaret.regression import *
import pandas as pd
```

```python
# read in data
# Download location - https://www.kaggle.com/yamaerenay/spotify-dataset-19212020-160k-tracks

spotify = pd.read_csv('data.csv')
```

```python
# using a sample of the dataset (you can use any amount)
spotify_sample = spotify.sample(1000)
spotify_sample.head()
```

```python
# setup your regression parameters 
regression = setup(data = spotify_sample, 
 target = 'popularity',
 session_id = 100, silent = True,
 ignore_features = ['id']
 )
```

```python
# compare models
compare_models()
```

```python
# create a model
catboost = create_model('catboost')
```

```python
# predict on test set
predictions = predict_model(catboost)
```

```python
# interpreting model
interpret_model(catboost)
```
